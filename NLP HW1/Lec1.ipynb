```python
from nltk.corpus import wordnet as wn
print(wn.synsets("mean")) # get the synonyms of "mean"
# The output includes synsets (sets of synonyms), such as “entail.v.01”:
# “entail” is the synonym.
# “v” indicates the part of speech (verb in this case).
# “01” is the sense number, as a word can have multiple meanings (senses).
```

```python
synsets = wn.synsets('mean') # get the synonyms of "mean"
for synset in synsets: # for loop for all the synsets
    print(f'synset:{synset.name()}') # get the full name of one synset
    print(f'Definition:{synset.definition()}') # get the definition of the word 
    print(f'Examples:{synset.examples()}') # get the example usage of the word
    print(f'POS:{synset.pos()}') # get the part-of-speech, 词性标注
    print()
```

```python
part_of_speech_mapping = {
    'n': 'noun',
    'v': 'verb',
    's': 'adj(s)',
    'a': 'adj',
    'r': 'adv'
} 
# define the POS mapping to the full name of pos

for synset in wn.synsets('mean'): # for loop for all the synsets of the word 'mean'
    
    pos_name = part_of_speech_mapping[synset.pos()]
    # get the pos full name of the current word

    lemma_names = [lemma.name() for lemma in synset.lemmas()]
    # get all lemma names(synonyms) for the current word

    lemma_names_str = ','.join(lemma_names) 
    #join the lemma_names into a comma-separated string

    print(f'{pos_name}:{lemma_names_str}')
```

```python
from nltk.corpus import wordnet as wn

dog = wn.synsets('dog')[0]
# get the first synset for the word dog

print(f'hypernyms:{dog.hypernyms()}')
for hypernym in dog.hypernyms():
    print(hypernym.name())
# get the hypernyms (more general terms)

print('\nHyponyms:')
for hyponym in dog.hyponyms():
    print(hyponym.name())
# get hyponym (more specific terms)
```

```python
import nltk
from nltk.tokenize import word_tokenize
#nltk.download('punkt_tab')
paragraphs = ['After completing my Master at UNC, my long-term goal is to pursue a Ph.D. in Statistics. The advanced training in statistical theory, machine learning, and stochastic modeling that UNC offers will provide the ideal foundation for my doctoral studies. ',
'I am particularly interested in contributing to the development of new statistical methodologies that address high-dimensional data problems, such as those encountered in healthcare and theory of statistical learning.',
'In the future, I hope to apply these methodologies to fields where robust statistical models can drive better decision-making and improve outcomes. ',
'A Ph.D. would allow me to influence the direction of research and develop new models that can address these real-world challenges rather than merely implementing existing solutions.',
'I look forward to the opportunity to be a part of UNC’s academic community. Thank you for considering my application.'
]

# 将段落转换为小写并进行分词
tokenized_paragraphs = [word_tokenize(para.lower()) for para in paragraphs]

print(tokenized_paragraphs)
```

```python
from gensim.models import Word2Vec
#Train a Word2Vec model using the above text via gensim.

model = Word2Vec(sentences=tokenized_paragraphs,
vector_size=100, window=5, min_count=1, workers=4) # define the vector size
```

```python
import numpy as np
word_vectors = model.wv
word = "machine"
embedding = model.wv[word]
print(f"Embedding for '{word}': {embedding}")
print(f"Shape of embedding: {np.array(embedding).shape}")
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import numpy as np

words = list(model.wv.index_to_key) # word vector back to word list
print(words)

vectors = [model.wv[word] for word in words] # map the word to its vector representation
tsne = TSNE(n_components=2, random_state=0, perplexity=5) 
vectors_2d = tsne.fit_transform(np.array(vectors)) # fit and transform the model on the given vector

plt.figure(figsize=(10, 8))
plt.scatter(vectors_2d[:, 0], vectors_2d[:, 1])
for i, word in enumerate(words):
    plt.annotate(word, xy=(vectors_2d[i, 0], vectors_2d[i, 1]))
plt.show()
```
